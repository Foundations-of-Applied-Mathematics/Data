```
import IPython.display
import PIL.Image
import Image
import numpy as np
import skimage.draw 
from random import randint

# A simple function to display an image in an ipython notebook
def nbimage( data ):
    IPython.display.display(PIL.Image.fromarray(data))
 
# Initialize 512*288 black 
image
image = np.zeros((512,288,3))
image = image.astype('uint8')

# Create 50 random shapes
# Create 25 random circles
for i in range(25):
  img = np.zeros((512, 288, 3), dtype=np.uint8)
  rr, cc = skimage.draw.circle(randint(50,450), randint(50,230), randint(1,50))
  image[rr, cc] = [randint(0,255), randint(0,255), randint(0,255)]

# Create 25 random lines
for i in range(25):
  img = np.zeros((512, 288, 3), dtype=np.uint8)
  rr, cc = skimage.draw.line(randint(0,511), randint(0,287), randint(0,511), randint(0,287))
  image[rr, cc] = [randint(0,255), randint(0,255), randint(0,255)]  
  
nbimage( image)               


```
